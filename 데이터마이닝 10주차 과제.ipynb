```python
import pandas as pd
from scipy import stats

df = pd.read_csv("branch_satisfaction.csv")

a = df[df["branch"] == "A"]["score"]
b = df[df["branch"] == "B"]["score"]
c = df[df["branch"] == "C"]["score"]

result = stats.f_oneway(a, b, c)

means = df.groupby("branch")["score"].mean()
print("지점별 평균:")
print(means)
print("F-statistic:", result.statistic)
print("p-value:", result.pvalue)
print("평균이 가장 높은 지점:", means.idxmax())
```

```python
import pandas as pd
from scipy import stats

df = pd.read_csv("before_after_worktime.csv")

before = df["before_min"]
after = df["after_min"]

diff = before - after

result = stats.ttest_rel(before, after, alternative="greater")

print("평균 차이(before - after):", diff.mean())
print("t-statistic:", result.statistic)
print("p-value:", result.pvalue)
```
